```python
import random # a dice with hundred size 1-50 house wins, 51-99 user wins, 100 house wins
import matplotlib.pyplot as plt
```

```python
def rolldice():
    roll=random.randint(1,100)
    if roll==100:
        #print('number is', roll, 'house wins , you loser. ')
        return False
    elif roll<=50:
        #print('number is', roll, 'house wins , you loser. ')
        return False
    else:
        #print('number is', roll, 'you win, lucky bastard. ')
        return True
    
def betting(inyourpocket,betmoney,howmanygames):
    moneylog=[]
    money1=inyourpocket
    money2=betmoney
    count=howmanygames
    for i in range(count):
        if rolldice():
            money1=money1+money2
            moneylog.append(money1)
        else:
            money1=money1-money2
            moneylog.append(money1)

    print('you have',money1,'left')
    
    plt.plot(moneylog)
```

```python
betting(10000,100,100)
```

```python
for i in range(100):
    betting(10000,100,10000)
```
