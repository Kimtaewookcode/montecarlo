```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
N=10000
x=np.random.uniform(low=-1,high=1, size=[N,1])
y=np.random.uniform(low=-1,high=1, size=[N,1])
circle=x**2+y**2<1
pi=4*np.sum(circle)/N
pi
```

```python
x_cir=x[circle]
y_cir=y[circle]
plt.scatter(x_cir,y_cir,s=1)
```
